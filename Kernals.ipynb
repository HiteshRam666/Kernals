###**Edge Detection Kernel**

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np

image_path = "/content/dg.jpg"
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

sobel_vertical = np.array([[+1, +2, +1],
                           [0, 0, 0],
                           [-1, -2, -1]])

sobel_output = cv2.filter2D(image, -1, sobel_vertical)
sobel_output.shape
```

```python
plt.figure(figsize=(10, 5))

# original image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap = 'gray')
plt.axis('off')

# Sobel Conv Edge Output
plt.subplot(1, 2, 2)
plt.title("Sobel Conv Edge Output")
plt.imshow(sobel_output, cmap='gray')
plt.axis('off')

plt.show()
```

###**Sharpening Kernel**

```python
sharp = np.array([[ 0, -1,  0],
                  [-1,  5, -1],
                  [ 0, -1,  0]])

sharp_kernel = cv2.filter2D(image, -1, sharp)

plt.figure(figsize=(10, 5))

# original image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap = 'gray')
plt.axis('off')

# Sharpening Output
plt.subplot(1, 2, 2)
plt.title("Sharpening Kernel Output")
plt.imshow(sharp_kernel, cmap='gray')
plt.axis('off')

plt.show()
```

```python
emboissing = np.array([[-2, -1, 0],
                      [-1,  1, 1],
                      [ 0,  1, 2]])

emb_kernel = cv2.filter2D(image, -1, emboissing)

plt.figure(figsize=(10, 5))

# original image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap = 'gray')
plt.axis('off')

# Sobel Conv Edge Output
plt.subplot(1, 2, 2)
plt.title("Emboissing Kernel Output")
plt.imshow(emb_kernel, cmap='gray')
plt.axis('off')

plt.show()
```
